```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
task2_1 = pd.read_csv("/content/task2_1.csv")
```

```python
task2_1
```

```python
# creating separate dataframes for Uber and Lyft
task2_1_uber = task2_1[task2_1["business"] == "Uber"]
task2_1_lyft = task2_1[task2_1["business"] == "Lyft"]
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.bar(data=task2_1_uber, x="month", height="trip_count", color='skyblue')  # Customize color if needed

# Adding title and labels
plt.title("Uber Trips per Month")
plt.xlabel("Month")
plt.ylabel("Trip Count")

# Displaying the plot
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.bar(data=task2_1_lyft, x="month", height="trip_count", color='skyblue')  # Customize color if needed

# Adding title and labels
plt.title("Lyft Trips per Month")
plt.xlabel("Month")
plt.ylabel("Trip Count")

# Displaying the plot
plt.show()
```
